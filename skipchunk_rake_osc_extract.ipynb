```python
import glob
import lib.skos
from lib.rake.rake_extract import rake_extract
from lib.skipchunk.skipchunk_extract import skipchunk_extract,sortgroup
```

```python
files = glob.glob("./content/osc/data/*.txt")
```

```python
skipchunk_concepts,skipchunk_predicates = skipchunk_extract(files)
```

```python
cons = sortgroup(skipchunk_concepts)
cnum = len(cons)
for c in cons[cnum-25:cnum][::-1]:
    print(skipchunk_concepts[c[1]][0][2])
```

```python
kos = lib.skos.SKOS()
baseuri = u'http://example.com/'
sckos = kos.scheme(baseuri + u'osc-scheme',u'OSC Scheme','osc-scheme')
for c in cons:
    conuri = baseuri + c[1]
    conobj = skipchunk_concepts[c[1]]
    concept = sckos.concept(conuri,c[1])
    pref = True
    for labels in conobj:
        label = labels[2]
        concept.label(label,u'en',isPref=pref)
        pref = False
print(sckos.turtle())
```

```python
preds = sortgroup(skipchunk_predicates)
spkos = kos.scheme(baseuri + u'osc-predicates',u'OSC Predicate Scheme','osc-predicates')
for c in preds:
    conuri = baseuri + c[1]
    conobj = skipchunk_predicates[c[1]]
    concept = spkos.concept(conuri,c[1])
    pref = True
    for labels in conobj:
        label = labels[2]
        concept.label(label,u'en',isPref=pref)
        pref = False
print(spkos.turtle())
```
